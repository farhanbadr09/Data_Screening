```python
import pandas as pd
empdata = pd.read_csv('MSFT.csv', index_col=False, delimiter = ',')
empdata.head()
```

```python
conn = mysql.connector.connect(host='localhost',
                                         database='employee',
                                         user='root',
                                         password='admin')
try:
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS farhan;')
        print('Creating table....')
        cursor.execute("CREATE TABLE farhan (Date Date , Open FLOAT(24) NOT NULL, High FLOAT(24) NOT NULL, Low FLOAT(24), Close FLOAT(24) NOT NULL, Adj_Close FLOAT(24), Volume FLOAT(24))")
        print("MSFT_Stocks Data table is created....")
        for i,row in empdata.iterrows():
            sql = "INSERT INTO farhan VALUES (%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not autocommitted by default, so we 
#              must commit to save our changes
            conn.commit()
except Error as e:
    print("Error while connecting to MySQL", e)
```

```python
# Execute query to SELCT * 
sql = "SELECT * FROM farhan"
cursor.execute(sql)

# Fetch all the records
result = cursor.fetchall()
for i in result:
    print(i)
```

```python
# import the mysql client library
try:
    # Obtain a cursor object
    cursor = conn.cursor()

    # Execute DESCRIBE statement
    cursor.execute("DESCRIBE farhan")

    # Fetch and print the meta-data of the table
    indexList = cursor.fetchall()

    print(indexList)

except Exception as e:
    print("Exception occured:{}".format(e))

finally:
    conn.close()
```

# 2. Calculation You are required to use this data and calculate the following using SQL

```python
# Weekly average of High, Low and Volume
try:
    # Obtain a cursor object
    cursor = conn.cursor()

    # Execute DESCRIBE statement
    cursor.execute("SELECT WEEK(Date),AVG(High),AVG(Low), AVG(Volume) FROM farhan group by week(Date);")

    # Fetch and print the meta-data of the table
    indexList = cursor.fetchall()

    print(indexList)

except Exception as e:
    print("Exception occured:{}".format(e))

finally:
    conn.close()
```

```python
# Monthly average of High, Low and Volume
try:
    # Obtain a cursor object
    cursor = conn.cursor()

    # Execute DESCRIBE statement
    cursor.execute("SELECT MONTH(Date),AVG(High),AVG(Low), AVG(Volume) FROM farhan group by MONTH(Date);")

    # Fetch and print the meta-data of the table
    indexList = cursor.fetchall()

    print(indexList)

except Exception as e:
    print("Exception occured:{}".format(e))

finally:
    conn.close()
```
